```python
from tkinter import*

## cửa sổ chọn hệ phương trình hay phương trình
window = Tk()
window.title("GIAO DIỆN GIẢI PT VÀ HPT")

def click_1():
    w1.grid_forget()
    wpt.grid(column = 1, row = 1)

def click_2():
    w1.grid_forget()
    whpt.grid(column = 1, row = 1)

def click_thoat():
    window.destroy()

w1 = Frame(window)
w1.grid(column = 1, row = 1)

### khởi tạo cửa sổ chọn phương trình hoặc hệ phương trình
lb1 = Label(w1, text = "CHỌN PHƯƠNG TRÌNH HOẶC HỆ PHƯƠNG TRÌNH", fg = "green", font = ("Bold Arial",15))
lb1.grid(column = 1, row = 1)

bt1 = Button(w1, text = "Phương trình", fg = "blue", font = ("Arial",15), command = click_1)
bt2 = Button(w1, text = "Hệ phương trình", fg = "blue", font = ("Arial",15), command = click_2)

bt1.grid(column = 1, row = 4,columnspan = 2,pady = (10,0))
bt2.grid(column = 1, row = 5,columnspan = 2,pady = (0,10))

bt_thoat = Button(w1, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
bt_thoat.grid(column = 1, row = 10)

## khởi tạo cửa sổ phương trình
wpt = Frame(window)

def click_1_1():
    wpt.grid_forget()
    wpt1.grid(column = 1, row = 1)

def click_1_2():
    wpt.grid_forget()
    wpt2.grid(column = 1, row = 1)
    
lb1 = Label(wpt, text = "CHỌN LOẠI PHƯƠNG TRÌNH",fg = "green", font = ("Arial",15))
lb1.grid(column = 1, row = 1)

bt1 = Button(wpt, text = "Phương trình bậc nhất", fg = "blue", font = ("Arial",15), command = click_1_1, width = 20)
bt2 = Button(wpt, text = "Phương trình bậc hai", fg = "blue", font = ("Arial",15), command = click_1_2, width = 20)

bt1.grid(column = 1, row = 4,columnspan = 2,pady = (10,0))
bt2.grid(column = 1, row = 5,columnspan = 2,pady = (0,10))

bt_thoat = Button(wpt, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
bt_thoat.grid(column = 1, row = 8)

## Phương trình bậc 1:
wpt1 = Frame(window)
lb1 = Label(wpt1, text = "GIẢI PHƯƠNG TRÌNH BẬC 1",fg = "green",font = ("Bold Arial",15))
lb1.grid(column = 1, row = 1,columnspan = 2)

## các thứ trong pt bậc 1 
lb2 = Label(wpt1, text = "Hệ số phương trình:", fg = "red", font = ("Arial",15))
lb2.grid(column = 1, row = 4)

a1_1 = StringVar()
b1_1 = StringVar()

# a
lb3 = Label(wpt1, text = "a =", fg = "red", font = ("Arial",15))
lb3.grid(column = 1, row = 8)

et1 = Entry(wpt1, font = ("Arial, 15"), width = 5, textvariable = a1_1)
et1.grid(column = 2, row = 8)

# b
lb4 = Label(wpt1, text = "b =", fg = "red", font = ("Arial",15))
lb4.grid(column = 1, row = 12)
et2 = Entry(wpt1, font = ("Arial, 15"), width = 5, textvariable = b1_1)
et2.grid(column = 2, row = 12)

def click_giai1():
    n1_1 = int(a1_1.get())
    n2_1 = int(b1_1.get())
    if n1_1 != 0:
        kq_pt_1 = ("Phương có một nghiệm duy nhất là "+ str(round((-n2_1/n1_1),2)))
    elif n2_1 != 0:
        kq_pt_1 = ("Phương trình vô nghiệm")
    else:
        kq_pt_1 = ("Phương trình có vô số nghiệm")
    
    lb_inpt_1 = Label(wpt1, text = ("Phương trình: " + str(n1_1)+ "x +" + " "+ str(n2_1) + " = 0" ), fg = "green", font = ("Arial", 15))
    lb_inpt_1.grid(column = 1, row = 18, columnspan = 3)
    lb5 = Label(wpt1, text = ("Kết quả: "+ str(kq_pt_1)),fg = "blue", font = ("Arial",15))
    lb5.grid(column = 1, row = 21, columnspan = 3)

def click_ql_1():
    wpt1.grid_forget()
    wpt = Frame(window)
    w1.grid(column = 1, row = 1)
    lb1 = Label(w1, text = "CHỌN PHƯƠNG TRÌNH HOẶC HỆ PHƯƠNG TRÌNH", fg = "green", font = ("Bold Arial",15), width = 40)
    lb1.grid(column = 1, row = 1)

    bt1 = Button(w1, text = "Phương trình", fg = "blue", font = ("Arial",15), command = click_1)
    bt2 = Button(w1, text = "Hệ phương trình", fg = "blue", font = ("Arial",15), command = click_2)

    bt1.grid(column = 1, row = 4,columnspan = 2,pady = (10,0))
    bt2.grid(column = 1, row = 5,columnspan = 2,pady = (0,10))

    bt_thoat = Button(w1, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
    bt_thoat.grid(column = 1, row = 10)
    
# nút giải:
bt_giai = Button(wpt1, text = "Giải", fg = "green", font = ("Arial", 15), command = click_giai1)
bt_giai.grid(column = 1, row = 15)
# nút thoát:
bt_thoat = Button(wpt1, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
bt_thoat.grid(column = 2, row = 24, columnspan = 2)
# nút quay lại:
bt_ql_1 = Button(wpt1, text = "Quay lại", fg = "red", font = ("Arial",15), command = click_ql_1, width = 7)
bt_ql_1.grid(column = 1, row = 24, columnspan = 1)


## Phương trình bậc 2:
wpt2 = Frame(window)
lb1 = Label(wpt2, text = "GIẢI PHƯƠNG TRÌNH BẬC 2",fg = "green",font = ("Bold Arial",15))
lb1.grid(column = 1, row = 2, columnspan = 3)

## các thứ ở trong
lb2 = Label(wpt2, text = "Hệ số phương trình:", fg = "red", font = ("Arial",15))
lb2.grid(column = 1, row = 4)

a1_2 = StringVar()
b1_2 = StringVar()
c1_2 = StringVar()

# a
lb3 = Label(wpt2, text = "a =", fg = "red", font = ("Arial",15))
lb3.grid(column = 1, row = 8)

et1 = Entry(wpt2, font = ("Arial, 15"), width = 5, textvariable = a1_2)
et1.grid(column = 2, row = 8)

# b
lb4 = Label(wpt2, text = "b =", fg = "red", font = ("Arial",15))
lb4.grid(column = 1, row = 12)
et2 = Entry(wpt2, font = ("Arial, 15"), width = 5, textvariable = b1_2)
et2.grid(column = 2, row = 12)

#c
lb5 = Label(wpt2, text = "c =", fg = "red", font = ("Arial",15))
lb5.grid(column = 1, row = 15)

et3 = Entry(wpt2, font = ("Arial, 15"), width = 5, textvariable = c1_2)
et3.grid(column = 2, row = 15)

def click_giai2():
    n1_2 = int(a1_2.get())
    n2_2 = int(b1_2.get())
    n3_2 = int(c1_2.get())
    denta_pt2 = ((n2_2)**2 - 4*(n1_2)*(n3_2))
    if n1_2 == 0:
        if n2_2 == 0 and n3_2 == 0:
            kq_pt_2 = ("Phương trình có vô số nghiệm")
        elif n2_2 == 0 and n3_2 !=0:
            kq_pt_2 = ("Phương trình vô nghiệm")
        else:
            kq_pt_2 = ("Phương trình có một nghiệm duy nhất x = "+ str(round(-n3_2/n2_2,2)))
    elif denta_pt2 < 0:
        kq_pt_2 = ("Phương trình vô nghiệm")
    elif denta_pt2 == 0:
        kq_pt_2 = ("Phương trình có nghiệm kép x1 = x2 = ", str(-n2_2/(2*n1_2)))
    else:
        x1 = ((-n2_2)+(denta_pt2)**0.5)/(2*n1_2)
        x2 = ((-n2_2)-(denta_pt2)**0.5)/(2*n1_2)

        kq_pt_2 = ("Phương trình có hai nghiệm phân biệt: x1 = "+ str(round(x1,2))+ " ; "+ "x2 = "+ str(round(x2,2)))
    
    lb_inpt_2 = Label(wpt2, text = ("Phương trình: "+ str(n1_2)+ "x^2 + "+ str(n2_2)+ "x + "+ str(n3_2)+ " "+ "= 0"), fg = "green", font = ("Arial", 15))
    lb_inpt_2.grid(column = 1, row = 21)
    lb6 = Label(wpt2, text = ("Kết quả: "+ str(kq_pt_2)),fg = "blue", font = ("Arial",15))
    lb6.grid(column = 1, row = 24)

def click_ql_2():
    wpt2.grid_forget()
    wpt = Frame(window)
    w1.grid(column = 1, row = 1)
    lb1 = Label(w1, text = "CHỌN PHƯƠNG TRÌNH HOẶC HỆ PHƯƠNG TRÌNH", fg = "green", font = ("Bold Arial",15), width = 40)
    lb1.grid(column = 1, row = 1)

    bt1 = Button(w1, text = "Phương trình", fg = "blue", font = ("Arial",15), command = click_1)
    bt2 = Button(w1, text = "Hệ phương trình", fg = "blue", font = ("Arial",15), command = click_2)

    bt1.grid(column = 1, row = 4,columnspan = 2,pady = (10,0))
    bt2.grid(column = 1, row = 5,columnspan = 2,pady = (0,10))

    bt_thoat = Button(w1, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
    bt_thoat.grid(column = 1, row = 10)    
    
# nút giải:
bt_giai = Button(wpt2, text = "Giải", fg = "green", font = ("Arial", 15), command = click_giai2)
bt_giai.grid(column = 1, row = 18)

# nút thoát
bt_thoat = Button(wpt2, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
bt_thoat.grid(column = 2, row = 27)

# nút quay lại:
bt_ql_2 = Button(wpt2, text = "Quay lại", fg = "red", font = ("Arial",15), command = click_ql_2, width = 7)
bt_ql_2.grid(column = 1, row = 27, columnspan = 1)

## khởi tạo cửa sổ hệ phương trình
whpt = Frame(window)

def click_hpt():
    whpt.grid_forget()
    whpt1.grid(column = 1, row = 1)
    
lb1 = Label(whpt, text = "CHỌN LOẠI HỆ PHƯƠNG TRÌNH",fg = "green",font = ("Arial",15))
lb1.grid(column = 1, row = 1, columnspan = 3)

bt1 = Button(whpt, text = "Hệ phương trình bậc nhất 2 ẩn", fg = "blue", font = ("Arial",15), command = click_hpt)
bt1.grid(column = 1, row = 4,columnspan = 2,pady = (10,0))

## Hệ phương trình 2 ẩn:
whpt1 = Frame(window)
lb1 = Label(whpt1, text = "GIẢI HỆ PHƯƠNG TRÌNH BẬC NHẤT 2 ẨN", fg = "green",font = ("Bold Arial",15))
lb1.grid(column = 1, row = 2, columnspan = 3)

## các thứ ở trong
lb2 = Label(whpt1, text = "Hệ số hệ phương trình:", fg = "red", font = ("Arial",15))
lb2.grid(column = 1, row = 4, columnspan = 2)

a1 = StringVar()
b1 = StringVar()
c1 = StringVar()
a2 = StringVar()
b2 = StringVar()
c2 = StringVar()

# a1
lb3 = Label(whpt1, text = "a1 =", fg = "red", font = ("Arial",15))
lb3.grid(column = 0, row = 9, columnspan = 1)

et1 = Entry(whpt1, font = ("Arial, 15"), width = 5, textvariable = a1)
et1.grid(column = 1, row = 9)

# b1
lb4 = Label(whpt1, text = "b1 =", fg = "red", font = ("Arial",15))
lb4.grid(column = 0, row = 12)
et2 = Entry(whpt1, font = ("Arial, 15"), width = 5, textvariable = b1)
et2.grid(column = 1, row = 12)

#c1
lb5 = Label(whpt1, text = "c1 =", fg = "red", font = ("Arial",15))
lb5.grid(column = 0, row = 15)

et3 = Entry(whpt1, font = ("Arial, 15"), width = 5, textvariable = c1)
et3.grid(column = 1, row = 15)

#a2
lb6 = Label(whpt1, text = "a2 =", fg = "red", font = ("Arial",15))
lb6.grid(column = 3, row = 9)

et4 = Entry(whpt1, font = ("Arial, 15"), width = 5, textvariable = a2)
et4.grid(column = 4, row = 9)

#b2
lb7 = Label(whpt1, text = "b2 =", fg = "red", font = ("Arial",15))
lb7.grid(column = 3, row = 12)

et5 = Entry(whpt1, font = ("Arial, 15"), width = 5, textvariable = b2)
et5.grid(column = 4, row = 12)

#c2
lb8 = Label(whpt1, text = "c2 =", fg = "red", font = ("Arial",15))
lb8.grid(column = 3, row = 15)

et6 = Entry(whpt1, font = ("Arial, 15"), width = 5, textvariable = c2)
et6.grid(column = 4, row = 15)

def click_giai3():
    n_a1 = int(a1.get())
    n_b1 = int(b1.get())
    n_c1 = int(c1.get())
    n_a2 = int(a2.get())
    n_b2 = int(b2.get())
    n_c2 = int(c2.get())
    D = n_a1*n_b2-n_a2*n_b1
    Dx = n_b2*n_c1-n_b1*n_c2
    Dy = n_a1*n_c2-n_a2*n_c1
    if D!=0:
        kq_hpt = "Hệ phương trình có nghiệm: x = "+ str(round(Dx/D,2)) + " và y = " + str(round(Dy/D,2))
    elif Dx ==0:
        kq_hpt = "Hệ phương trình có vô số nghiệm"
    else:
        kq_hpt = "Hệ phương trình vô nghiệm"
    
    lb_kq_hpt = Label(whpt1, text = ("Kết quả: "+ str(kq_hpt)), fg = "blue", font = ("Arial",15))
    lb_kq_hpt.grid(column = 1, row = 18, columnspan = 3)
    
def click_ql_3():
    whpt1.grid_forget()
    wpt = Frame(window)
    w1.grid(column = 1, row = 1)
    lb1 = Label(w1, text = "CHỌN PHƯƠNG TRÌNH HOẶC HỆ PHƯƠNG TRÌNH", fg = "green", font = ("Bold Arial",15), width = 40)
    lb1.grid(column = 1, row = 1)

    bt1 = Button(w1, text = "Phương trình", fg = "blue", font = ("Arial",15), command = click_1)
    bt2 = Button(w1, text = "Hệ phương trình", fg = "blue", font = ("Arial",15), command = click_2)

    bt1.grid(column = 1, row = 4,columnspan = 2,pady = (10,0))
    bt2.grid(column = 1, row = 5,columnspan = 2,pady = (0,10))

    bt_thoat = Button(w1, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
    bt_thoat.grid(column = 1, row = 10)    

# nút giải:
bt_giai = Button(whpt1, text = "Giải", fg = "green", font = ("Arial", 15), command = click_giai3)
bt_giai.grid(column = 1, row = 27)

# nút thoát
bt_thoat = Button(whpt1, text = "Thoát", fg = "red", font = ("Arial",15), command = click_thoat, width = 5)
bt_thoat.grid(column = 2, row = 30, columnspan = 3)

# nút quay lại:
bt_ql_2 = Button(whpt1, text = "Quay lại", fg = "red", font = ("Arial",15), command = click_ql_3, width = 7)
bt_ql_2.grid(column = 1, row = 30, columnspan = 1)

window.mainloop()
```
